```python

#import library
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = "citra.jpg"
img = cv2.imread(image_path)

# Salin citra ke variabel khusus
img_copy = img.copy()

# Dapatkan atribut khusus citra (ukuran)
height, width = img_copy.shape[:2]

# Tampilkan atribut khusus
print(f"Ukuran citra: {width} x {height}")

# 1. Citra Asli
original = img_copy

# 2. Rotated
(h, w) = img_copy.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)  # Rotasi 45 derajat
rotated = cv2.warpAffine(img_copy, M, (w, h))

# 3. Resized
resized = cv2.resize(img_copy, (w // 2, h // 2))  # Ubah ukuran menjadi setengah dari ukuran asli

# 4. Cropped
cropped = img_copy[0:h // 2, 0:w // 2]  # Potong setengah bagian kiri atas gambar

# 5. Flipped
flipped = cv2.flip(img_copy, 0)  # Balik vertikal

# 6. Translated (digeser)
M = np.float32([[1, 0, 50], [0, 1, 50]])  # Geser 50 piksel ke kanan dan 50 piksel ke bawah
translated = cv2.warpAffine(img_copy, M, (w, h))

# Convert BGR to RGB for matplotlib
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
rotated = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)
resized = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
cropped = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)
flipped = cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB)
translated = cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)

# Display the images with their attributes
plt.figure(figsize=(15, 10))

# Citra Asli
plt.subplot(2, 3, 1)
plt.imshow(original)
plt.title('Citra Asli')
plt.xlim(0, width)
plt.ylim(height, 0)


# Rotated
plt.subplot(2, 3, 2)
plt.imshow(rotated)
plt.title('After Rotated')
plt.xlim(0, rotated.shape[1])
plt.ylim(rotated.shape[0], 0)

# Resized
plt.subplot(2, 3, 3)
plt.imshow(resized)
plt.title('After Resized')
plt.xlim(0, resized.shape[1])
plt.ylim(resized.shape[0], 0)

# Cropped
plt.subplot(2, 3, 4)
plt.imshow(cropped)
plt.title('After Cropped')
plt.xlim(0, cropped.shape[1])
plt.ylim(cropped.shape[0], 0)

# Flipped
plt.subplot(2, 3, 5)
plt.imshow(flipped)
plt.title('After Flipped')
plt.xlim(0, flipped.shape[1])
plt.ylim(flipped.shape[0], 0)

# Translated
plt.subplot(2, 3, 6)
plt.imshow(translated)
plt.title('After Translated')
plt.xlim(0, translated.shape[1])
plt.ylim(translated.shape[0], 0)

plt.tight_layout()
plt.show()
```
